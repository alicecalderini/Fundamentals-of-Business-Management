```python
import pandas as pd
```

```python
df = pd.read_csv("master_dataset.csv")
```

```python
print(df.columns)
```

# Cleaning: dropping double X6

```python
df = df.drop(["X6_Disaster_Event_Count"], axis=1)
```

# Renaming columns

```python
#Renaming columns
df = df.rename(columns={"Inventory_Turnover": "X12_Inventory_Turnover","X12_Economic_Policy_Uncertainty": "X11_Economic_Policy_Uncertainty", "X_PPI_All_Commodities" : "X13_PPI_All_Commodities"})
```

# Moving columns in the right position

```python
col = "X13_PPI_All_Commodities"
df = df[[c for c in df.columns if c != col] + [col]] #moving X13 in the last position
```

```python
#moving X6

col = "X6_Disaster_Cost_Sum"
new_position = 24

cols = df.columns.tolist()
cols.insert(new_position, cols.pop(cols.index(col)))

df = df[cols]
```

```python
#Moving X9
col = "X9_US_Dollar_Index"
new_position = 26

cols = df.columns.tolist()
cols.insert(new_position, cols.pop(cols.index(col)))

df = df[cols]
```

```python
#Moving X12
col = "X12_Inventory_Turnover"
new_position = 28

cols = df.columns.tolist()
cols.insert(new_position, cols.pop(cols.index(col)))

df = df[cols]
```

```python
print(df.columns) #checking if columns are in the right position
```

# Adding missing feature X8

```python
x_8 = pd.read_csv("quarterly_data/US_Scrap_Exports_quarterly.csv")

df_merged = pd.merge(
    df,
    x_8,
    on="QuarterEnd",    # common column to merge on
    how="left"       # type of merge (left: keep all rows from df)
)
```

```python
df_merged = df_merged.rename(columns={"US_Scrap_Exports": "X8_US_Scrap_Exports"}) #rename X8
```

```python
#moving X8 to the right position
col = "X8_US_Scrap_Exports"
new_position = 25

cols = df_merged.columns.tolist()
cols.insert(new_position, cols.pop(cols.index(col)))

df_merged = df_merged[cols]
```

```python
print(df_merged.columns) #checking if everything's ok
```

# Creating master dataset

```python
df_merged.to_csv("master_dataset_general.csv", index=False)
```
